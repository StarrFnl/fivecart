```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = data['count']
y = data['length']
```

```python
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1), y)
```

```python
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show()
```
