```python
classic_100 = pd.read_csv("./2023년도 성균 고전 100선.csv", encoding='cp949')
classic_100
```

```python
classic_100 = classic_100.drop_duplicates(subset='ISBN',keep='first')
classic_100
```

```python
import re

pattern = r'\([^)]*\)'
patternKor = r'[^ㄱ-ㅣ가-힣\s]'
for idx in classic_100.index:
    classic_100.loc[idx, '제목'] = re.sub(pattern=pattern, repl='', string= classic_100.loc[idx, '제목'])
    classic_100.loc[idx, '제목'] = re.sub(pattern=patternKor, repl='', string= classic_100.loc[idx, '제목'])

classic_100
```

```python
classic_100 = classic_100.drop_duplicates(subset='제목',keep='first')
classic_100
```

```python
classic_100.to_csv('classic_drop_dup.csv',index=False, encoding='utf-8')
```
