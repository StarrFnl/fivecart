```python
import numpy as np
import pandas as pd
import itertools
```

```python
data_2022 = pd.read_csv("key_2022.csv", encoding='cp949')
data_2021 = pd.read_csv("key_2021.csv", encoding='cp949')
data_2020 = pd.read_csv("key_2020.csv", encoding='cp949')
data_2019 = pd.read_csv("key_2019.csv", encoding='cp949')
data_2018 = pd.read_csv("key_2018.csv", encoding='cp949')
data_2017 = pd.read_csv("key_2017.csv", encoding='cp949')
```

```python
data_2022 = data_2022.drop(['Unnamed: 0.1'], axis=1)
data_2021 = data_2021.drop(['Unnamed: 0.1'], axis=1)
data_2020 = data_2020.drop(['Unnamed: 0.1'], axis=1)
data_2019 = data_2019.drop(['Unnamed: 0.1'], axis=1)
data_2018 = data_2018.drop(['Unnamed: 0.1'], axis=1)
data_2017 = data_2017.drop(['Unnamed: 0.1'], axis=1)
```

```python
result1 = pd.concat([data_2022,data_2021,data_2020, data_2019, data_2018, data_2017], ignore_index=True)
```
