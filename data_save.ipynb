```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./fivecart_reports_new_one.csv", encoding='utf-8')
data
```

```python
data_2022 = data.loc[(data['date_writer'].str.startswith('2022'))] 
data_2022
```

```python
data_2023 = data.loc[(data['date_writer'].str.startswith('2023'))] 
data_2023
```

```python
data_2023.to_csv('fivecart_reports_2023.csv',index=False, encoding='utf-8')
```

```python
data_2022_old = pd.read_csv("./fivecart_reports_2022.csv", encoding='cp949')
```

```python
data_2022_x = pd.concat([data_2022, data_2022_old])
data_2022_x
```

```python
data_2022_x.reset_index(inplace=True)
data_2022_x
```

```python
data_2022_x = data_2022_x.drop(columns = ['index'])
data_2022_x
```

```python
data_2022_x.to_csv('fivecart_reports_2022.csv',index=False, encoding='utf-8')
```
