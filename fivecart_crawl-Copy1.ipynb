```python
import requests
from lxml import html
first_url = "https://book.skku.edu/read-reviews/review/page/1/" # 크롤링 원하는 주소 first page xpath
response = requests.get(first_url) #first page 진입
response 
last_url = html.fromstring(response.text).xpath('//*[@id="post-19"]/div[5]/div/div/div[2]/div/ul/li[12]/a/@href')#오거서 Last page xpath 
last_url
```

```python
import re
# last_num = re.sub(r'[^0-9]*', '', last_url[0]) #last page page번호만 남기기(다른 기호들 포함되어 긁어짐)
last_num = 83

book_urls = []    

# 끝까지 긁어지는 코드 1부터 last page까지 방문
for i in range(1, int(last_num)+1):
    print(i, end="|")
    fivecart_list_page = "https://book.skku.edu/read-reviews/review/page/"+str(i)+"/"
#첫 2개만 긁는 코드
# for i in range(1, 3):
#     fivecart_list_page = "https://book.skku.edu/read-reviews/review/page/"+str(i)+"/"
    response = requests.get(fivecart_list_page)#각 페이지 진입
#     response
    #페이지에서 독후감들 링크 추출, 저장
    book_urls = book_urls + [i for i in html.fromstring(response.text).xpath('//*[@id="post-19"]/div[5]/div/div/div[1]/ul/li/div/div/div[2]/h3/a/@href')]
    
    
print(len(book_urls))
```

```python
all_reports = []
not_crawled = []
cnt = 0
#링크 리스트에서 각 독후감 방문
for book_url in book_urls:
    
    cnt = cnt+1
    print(cnt,end="|")
    book_page = requests.get(book_url)
    #id 가져오기. 메타데이터 xpath id로 구분되는듯
    article_id = html.fromstring(book_page.text).xpath('//*[@id="print-div"]/div/article/@id')
    
    #각 id 붙여서 탐색
    writer_path = '//*[@id="'+article_id[0]+'"]/div/div/div/div[1]/div/div[2]/div[1]/div[2]/div[1]' #독후감 글쓴이 path
    report_writer = [i.text for i in html.fromstring(book_page.text).xpath(writer_path)][0]
    date_path = '//*[@id="'+article_id[0]+'"]/div/div/div/div[1]/div/div[2]/div[1]/div[2]/div[2]' # 날짜 path
    date_writer = [i.text for i in html.fromstring(book_page.text).xpath(date_path)][0]
    title_path = '//*[@id="'+article_id[0]+'"]/div/div/div/div[1]/div/div[2]/h3' # 제목 path
    title = [i.text for i in html.fromstring(book_page.text).xpath(title_path)][0]
    triple_path = '//*[@id="'+article_id[0]+'"]/div/div/div/div[1]/div/div[2]/div[2]/span/text()' #작가, 출판사, 출판일 path 
    triple = [i for i in html.fromstring(book_page.text).xpath(triple_path)]
    if(len(triple)!= 3):
        cnt = cnt-1
        not_crawled.append(book_url)
        continue
    book_writer = triple[0] #작가
    book_company = triple[1] #출판사
    book_date = triple[2] #출판일
    
    text_path = '//*[@id="'+article_id[0]+'"]/div/div/div/div[2]//text()' # 본문 path, 부모클래스 single_contbx 자식 텍스트 가져옴
    report_text = html.fromstring(book_page.text).xpath(text_path)
    report_text = "".join(report_text) #리스트 형식 결과물 join
    report_text = re.sub('(([\r&\t&\n&\xa0])\\2{1,})', '', report_text) #중복되어 나오는 특수문자들 1회만 나오게 변경
    report_text = report_text.replace("\r\n", " ").replace("\r", " ").replace("\n", " ").replace("\t", " ").replace("\xa0", " ") #공백 특수문자들 스페이스바로 대체
    
    all_reports.append({'report_writer':report_writer, 'date_writer':date_writer, 'title':title, 'book_writer':book_writer, 'book_company':book_company, 'book_date':book_date, 'report_text':report_text})
```
